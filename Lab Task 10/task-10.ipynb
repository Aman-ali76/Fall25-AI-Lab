```python
def fillnaObjectMode(cols):
    for i in cols:
        df[i].fillna(df[i].mode()[0] , inplace=True)

columns = ['Gender', 'Major', 'PartTimeJob', 'ExtraCurricularActivities']
fillnaObjectMode(columns)
```

```python
def fillnaFloatMode(cols):
    for i in cols:
        df[i].fillna(df[i].median(), inplace=True)

columns = ['Age','StudentID','StudyHoursPerWeek','AttendanceRate']
fillnaFloatMode(columns)
```

```python
def convertObjToInt(cols):
    for i in cols:
        data = pd.DataFrame({i:df[i].unique()})
        dat_LE = enc()
        dat_LE.fit(np.ravel(data))
        df[i] = dat_LE.transform(df[i])
        
columns = ['Gender', 'Major', 'PartTimeJob','ExtraCurricularActivities']
convertObjToInt(columns)
```
