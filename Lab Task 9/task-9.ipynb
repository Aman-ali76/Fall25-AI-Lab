```python
def mean_col(col):
    return df[col].mean()

def median_col(col):
    return df[col].median()

def mod_col(col):
    return df[col].mode()[0]

```

```python
print("Mean of Bathrooms is  : ", mean_col("bathrooms"))
print("Median of Bathrooms is  : ", median_col("bathrooms"))
print("Mode of Bathrooms is  : ", mod_col("bathrooms"))
```
