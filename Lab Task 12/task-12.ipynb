<div class="alert alert-block alert-warning">
    4.1. <b>Manage Null Values</b>
</div>

```python
df.isnull().sum()
```

```python
def fillnaObjectMode(cols):
    for i in cols:
        df[i].fillna(df[i].mode()[0] , inplace=True)

columns = ['city', 'street', 'statezip']
fillnaObjectMode(columns)
```

```python
def fillnaIntMode(cols):
    for i in cols:
        df[i].fillna(df[i].mode()[0], inplace=True)

columns = ['bedrooms','bathrooms','floors','waterfront','view','yr_built','yr_renovated']
fillnaIntMode(columns)
```

```python

def fillnaFloatMean(cols):
    for i in cols:
        df[i].fillna(df[i].mean(), inplace=True)

columns = ['price','condition','sqft_living','sqft_lot','sqft_above','sqft_basement']
fillnaFloatMean(columns)
```

<div class="alert alert-block alert-warning">
    4.2. <b>Manage Data Types</b>
</div>

```python

def ChangeFloatToInt(cols):
    for i in cols:
        df[i] = df[i].astype('int64')


columns = ['price','bathrooms','sqft_living','sqft_lot' , 'floors' , 'waterfront','view','condition','sqft_above','sqft_basement','yr_built','yr_renovated']

ChangeFloatToInt(columns)
```

```python
df.drop(columns=['date','country'],inplace=True)
```

```python
df.info()
```

```python
def convertObjToInt(cols):
    for i in cols:
        data = pd.DataFrame({i:df[i].unique()})
        dat_LE = enc()
        dat_LE.fit(np.ravel(data))
        df[i] = dat_LE.transform(df[i])
        
columns = ['street', 'city', 'statezip']
convertObjToInt(columns)
convertObjToInt(columns)
```

```python
df.info()
```

<div class="alert alert-block alert-info">
    5. <b>Train Test Split</b>
</div>

```python
X = df.drop('price', axis=1)   
y = df['price']                
```

```python
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size=0.3, shuffle=False)
```

```python
print(train_X.shape, train_y.shape)
print(test_X.shape, test_y.shape)
```

<div class="alert alert-block alert-info">
    6. <b>Applying Classifier / Model Training</b>
</div>

```python

model_svc = SVC()
model_svc.fit(train_X, train_y)

print(model_svc)
```

```python
model_pred_svc = model_svc.predict(test_X)
model_pred_svc
```

```python

model_acc_svc = accuracy_score(test_y, model_pred_svc)
print(round(model_acc_svc,3)*100, "%")
```
